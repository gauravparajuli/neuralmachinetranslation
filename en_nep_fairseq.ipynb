# 1 Training Data Preparation

To train a new model we need few elements/steps:

1.   Parallel corpus
2.   Sentencepiece model
3.   Data pre-processing





### Parallel corpus

```python
# prompt: count number of entries in train.en and train.ne and verify that number of lines are equal

!wc -l train.en
!wc -l train.ne
```

```python
# prompt: both train.en and train.ne contains 151937 lines.
# each line in train.en has its corresponding nepali translation in train.ne in the same line
# divide train.en and train.ne into training and validation set in the ratio of 8:2
# new files should be names as train.en, train.ne, valid.en, valid.ne

import random

random.seed(42) # set the seed for reproducibility

# Read the files
with open('train.en', 'r') as f_en, open('train.ne', 'r') as f_ne:
    lines_en = f_en.readlines()
    lines_ne = f_ne.readlines()

# Combine and shuffle the lines
combined_lines = list(zip(lines_en, lines_ne))
random.shuffle(combined_lines)

# Calculate split points
total_lines = len(combined_lines)
train_split = int(0.8 * total_lines)

# Split the data
train_data = combined_lines[:train_split]
valid_data = combined_lines[train_split:]

# Write the new files
with open('train.en', 'w') as f_train_en, open('train.ne', 'w') as f_train_ne:
    for en, ne in train_data:
        f_train_en.write(en)
        f_train_ne.write(ne)

with open('valid.en', 'w') as f_valid_en, open('valid.ne', 'w') as f_valid_ne:
    for en, ne in valid_data:
        f_valid_en.write(en)
        f_valid_ne.write(ne)
```

### Dataset pre-processing

Here, we have 2 tasks to do:

1.   Using the Sentencepiece model to pre-process our data (we would also do this to validation/test sets!). It will split the sentences into **subwords**, adding the special symbol `▁` to the first subword of a word.
2.   (Partially optional) Binarize the data to the Fairseq format.

Second step is partially optional as we can work with the raw data by providing different than default flag with `--dataset-impl=raw`.

Partially - as we have to run it anyway, but it won't binarize the data anymore.

```python
# english to nepali translation model
!fairseq-train \
        "ne2en20000" \
        --fp16 \
        --arch transformer_wmt_en_de \
        --share-all-embeddings \
        --optimizer adam \
        --adam-betas '(0.9, 0.98)' \
        --clip-norm 0.0 \
        --lr 5e-4 \
        --lr-scheduler inverse_sqrt \
        --warmup-updates 4000 \
        --warmup-init-lr 1e-07 \
        --dropout 0.1 \
        --weight-decay 0.0 \
        --criterion label_smoothed_cross_entropy \
        --label-smoothing 0.1 \
        --save-dir "model_output" \
        --log-format json \
        --log-interval 100 \
        --max-epoch 50 \
        --batch-size 80 \
        --patience 5 \
        --seed 3921 \
        --eval-bleu \
        --eval-bleu-args '{"beam": 5, "max_len_a": 1.2, "max_len_b": 10}' \
        --eval-bleu-detok space \
        --eval-bleu-remove-bpe sentencepiece \
        --eval-bleu-print-samples \
        --best-checkpoint-metric bleu \
        --maximize-best-checkpoint-metric \
        --no-epoch-checkpoints \
        --wandb-project "OLEN NMT" \
        # --max-tokens 8000
wandb.finish()
```
